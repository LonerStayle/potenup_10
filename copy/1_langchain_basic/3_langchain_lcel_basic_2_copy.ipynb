### 체인 만들기

```python
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
template = """
최근 {year}년 간 {field} 분야의 기술 발전에 대해서 알려줘
"""
prompt_template = PromptTemplate.from_template(template)

model = ChatOpenAI(
    temperature=0.1,
    model="gpt-4.1-mini",
    verbose=True
)

outputparser = StrOutputParser()

chain = prompt_template | model | outputparser
chain
```

1. invoke 출력
2. stream 출력
3. batch 출력
4. ainvoke
5. astream 비동기 출력

```python
#배치 단위 출력
response = chain.batch([{"field": "AI", "year": 3}, {"field": "드론", "year": 3}])
print(response)
```

```python
for i in response:
    print(i)
    print("-"*50)
```

```python
async_answer = chain.ainvoke({"field":"AI", "year":3})
```

```python
result = await async_answer
result
```

```python
async_stream = chain.astream({"field":"AI", "year":3})
async for token in async_stream:
    print(token, end="", flush=True)
```
